# Function to calculate weight slab:


- based on given table function is:

```python
def weight_slab(weight):
    i = round(weight % 1, 1)
    if i == 0.0:
        return weight
    elif i > 0.5:
        return int(weight) + 1.0
    else:
        return int(weight) + 0.5
```

# To apply weight slab to the columns:

```python
Company_X['Weight slab as per X (KG)'] = (Company_X['Total weight as per X (KG)']).apply(weight_slab)
Company_X.columns
```

# Computation of Total Expected Charge as per X:

```python
for _, row in Company_X.iterrows():
    fwd_category = 'fwd_' +row['Delivery Zone as per X']
    print(Courier_Company_Rates.at[0, fwd_category + '_fixed'])
```

```python
# list to store the expected charges
total_Expected_charge = []

# Loop through each row in Company_X dataset
for i in range(len(Company_X)):
    # Determine the forward and RTO charge categories based on the Delivery Zone
    delivery_zone = Company_X['Delivery Zone as per X'][i]
    
    # Fetch fixed and additional charges for forward and RTO based on the zone
    fwd_fixed = Courier_Company_Rates.loc[Courier_Company_Rates['Zone'] == delivery_zone, 'Forward Fixed Charge'].values[0]
    fwd_additional = Courier_Company_Rates.loc[Courier_Company_Rates['Zone'] == delivery_zone, 'Forward Additional Weight Slab Charge'].values[0]
    rto_fixed = Courier_Company_Rates.loc[Courier_Company_Rates['Zone'] == delivery_zone, 'RTO Fixed Charge'].values[0]
    rto_additional = Courier_Company_Rates.loc[Courier_Company_Rates['Zone'] == delivery_zone, 'RTO Additional Weight Slab Charge'].values[0]

    # Calculate the charges based on the Type of Shipment and Weight Slab
    if Company_X['Type of Shipment'][i] == 'Forward charges':
        if Company_X['Weight slab as per X (KG)'][i] <= 0.5:
            total_Expected_charge.append(fwd_fixed)
        else:
            additional_weight = (Company_X['Weight slab as per X (KG)'][i] - 0.5) / 0.5
            total_Expected_charge.append(fwd_fixed + additional_weight * fwd_additional)
    elif Company_X['Type of Shipment'][i] == 'Forward and RTO charges':
        if Company_X['Weight slab as per X (KG)'][i] <= 0.5:
            total_Expected_charge.append(fwd_fixed + rto_fixed)
        else:
            additional_weight = (Company_X['Weight slab as per X (KG)'][i] - 0.5) / 0.5
            total_Expected_charge.append(fwd_fixed + additional_weight * (fwd_additional + rto_additional))

# Add the calculated expected charges to the DataFrame
Company_X['Expected Charge as per X (Rs.)'] = total_Expected_charge

# Display the updated DataFrame
Company_X.head(5)
```
